```python
img = cv2.imread('image.jpg', cv2.IMREAD_COLOR)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
height, width, channels = img.shape


print('이미지 너비: ', width)
print('이미지 높이: ', height)
print('이미지 채널 수: ', channels)
```

```python
resized_img = cv2.resize(img, (300, 200))

print('변경된 이미지 높이: ', resized_img.shape[0])
print('변경된 이미지 너비: ', resized_img.shape[1])

plt.imshow(cv2.cvtColor(resized_img, cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(cv2.cvtColor(resized_img, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.savefig('resized_image.png', bbox_inches='tight', pad_inches=0)

FileLink('resized_image.png')
```

```python
img = cv2.imread('image.jpg', cv2.IMREAD_COLOR)
rotated_img = cv2.rotate(img,cv2.ROTATE_90_CLOCKWISE)
plt.imshow(cv2.cvtColor(rotated_img, cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(cv2.cvtColor(rotated_img, cv2.COLOR_BGR2RGB))
plt.axis('off')  # 축을 제거합니다.

plt.savefig('rotated_image.png', bbox_inches='tight', pad_inches=0)

FileLink('rotated_image.png')
```

```python
import cv2
import tkinter as tk
from tkinter import filedialog
from PIL import Image, ImageTk

def display_image(img, label):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    im = Image.fromarray(img)
    imgtk = ImageTk.PhotoImage(image=im) 
    label.config(image=imgtk)
    label.image = imgtk

def resize_image(img):
    resized_img = cv2.resize(img, (300, 200))
    return resized_img

def rotate_image(img):
    (h, w) = img.shape[:2]
    center = (w / 2, h / 2)
    M = cv2.getRotationMatrix2D(center, 90, 1.0)
    rotated_img = cv2.warpAffine(img, M, (w, h))
    return rotated_img

root = tk.Tk()

frame = tk.Frame(root)
frame.pack()

img_label = tk.Label(frame)
img_label.pack()

resize_button = tk.Button(frame, text="Resize Image", command=lambda: display_image(resize_image(upload_image()), img_label))
resize_button.pack()

rotate_button = tk.Button(frame, text="Rotate Image", command=lambda: display_image(rotate_image(upload_image()), img_label))
rotate_button.pack()

root.mainloop()
```
